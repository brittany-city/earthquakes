```python
import requests
```

```python
import pandas as pd
```

```python
#import url + 'query' to the end
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
```

```python
#collect data from a certain timeframe
payload = {'format':'csv',
          'starttime':'1990-01-01',
          'minlatitude': '34.946669',
          'maxlatitude': '36.533847',
          'minlongitude': '-90.229141',
          'maxlongitude': '-81.663552'}
```

```python
#r means response
r = requests.get(url=url, params=payload)
```
